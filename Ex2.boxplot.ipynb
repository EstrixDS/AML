```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn
import pandas as pd
np.random.seed(1)
```

```python
# multiply and add by random numbers to get some real values
# randn generates samples from the normal distribution (important - see below)
random_data = np.random.randn(50000)  * 20 + 20
```

```python
sns.boxplot(data=random_data)
plt.show()
```

```python
df1 = pd.read_csv('data/loans_income.csv')
data1 = df1['x'].to_numpy()
sns.boxplot(data=data1)
plt.show()
```

```python
df2 = pd.read_csv('data/SOCR-HeightWeight.csv')
data2 = df2['Height(Inches)'].to_numpy()
sns.boxplot(data=data2)
plt.show()
```

```python
df3 = pd.read_csv("https://userpage.fu-berlin.de/soga/200/2010_data_sets/students.csv")
data3 = df3[df3['gender'] == 'Female']["height"].to_numpy()
sns.boxplot(data=data3)
plt.show()
```

## numpy randn

since the function randn gives normal distributed values, the boxplot shows a perfect example of how a normal distributed dataset looks in a boxplot: The "minimum", the 1st Quartile being the lower boundary of the colored box, the median being the middle line of the coloured box, the 3rd quartile being the top boundary of the colored box and the "maximum". The minimum and maximum are defined as being 1.5 times the Interquartile Range subtracted by the line of the first Quartile and the maximum being defined as 1.5 times the Interquartile Range added to the line of the third Quartile. Therefore the outliers are the datapoints, which are further away than the minimum or maximum, in relation to the median.

## Loans income

The median is not in the middle of the box and all outliers are over the maximum. It does not only visualize, wether the dataset is normal distributed or not, but also display the skewness and kurtosis. Skewness is a measure of symmetry, or to be precise, the lack of symmetry. Kurtosis is a measure of wether the dataset is heavy-tailed, meaning a high standard deviation and/or lots of outliers, or wether the dataset is light-tailed, meaning small standard deviation or low amount of outliers.

## SOCR Height

The boxplot shows a median, which is in the middle of the box, and outliers on both sides. This shows that the data is, like the numpy randn data, normal distributed.

## students female height

The boxplot shows a median, which is a little bit closer to the 3rd quartile than to the first quartile. Also the amount of outliers are way less, but that has to do with the amount of data. Since this dataset is way smaller than the other ones, we cannot automatically say that this data has a lower kurtosis than the others.
